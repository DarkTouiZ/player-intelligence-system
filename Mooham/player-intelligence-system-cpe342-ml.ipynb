# Task 2 and 3 Equality Value Check

```python
Temp = pd.merge(
    task2[Dub_cols_23].drop("id", axis = 1),
    task3[Dub_cols_23].drop("id", axis = 1),
    on = "player_id",
    how = "inner",
    suffixes = ("_task2", "_task3")
)
```

```python
Dub_cols = [x for x in Dub_cols_23 if x not in ["id", "player_id"]]

Comparison_Results = {}

for col in Dub_cols:
    Equal_mask = Temp[f"{col}_task2"] == Temp[f"{col}_task3"]
    # Allow for NaN == NaN (treat both missing as equal)
    Equal_mask = Equal_mask | (Temp[f"{col}_task2"].isna() & Temp[f"{col}_task3"].isna())
    
    Match_Ratio = Equal_mask.mean()
    Comparison_Results[col] = round(Match_Ratio * 100, 2)

pd.DataFrame(list(Comparison_Results.items()), columns = ["Column", "Percent_Equal"])
```

## Modeling on join inner Test

### Baseline

```python
import numpy as np
from sklearn.preprocessing import OneHotEncoder

df = task2.drop(columns = ["id", "player_id"])
df.drop(columns = ["random_metric_1", "random_metric_2", "random_metric_3"], inplace = True)

ordinal_cols  = {
    "engagement_level": {"Low": 0, "Mid": 1, "High": 2, "Veteran" : 3},
    "loyalty_tier": {"Bronze": 0, "Silver": 1, "Gold": 2, "Platinum" : 3, "Diamond" : 4},
    "skill_tier" : {"Bronze": 0, "Silver": 1, "Gold": 2, "Platinum" : 3, "Diamond" : 4}
}

object_cols = df.select_dtypes(include = "object").columns
nominal_cols = [c for c in object_cols if c not in list(ordinal_cols.keys())]

numeric_cols = df.select_dtypes(include = [np.number]).columns.tolist()
numeric_cols.remove("segment")

# Encode ordinal features
df_ord = pd.DataFrame()
for col, mapping in ordinal_cols.items():
    df_ord[col] = df[col].map(mapping)

# Encode nominal features
ohe = OneHotEncoder(sparse_output = False, handle_unknown = "ignore")

df_nom = pd.DataFrame(
    ohe.fit_transform(df[nominal_cols]),
    columns = ohe.get_feature_names_out(nominal_cols),
    index = df.index
)
```

```python
import lightgbm as lgb

X = pd.concat([df[numeric_cols], df_ord, df_nom], axis = 1)
y = df["segment"]

Classifier = lgb.LGBMClassifier(
    objective = "multiclass",
    num_class = 4,
    random_state = 42,
)
```

```python
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.metrics import make_scorer, f1_score

Stratified_K_Fold = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)
Scorer = make_scorer(f1_score, average = "macro")

# Run CV
Scores = cross_val_score(
    Classifier,
    X,
    y,
    cv = Stratified_K_Fold,
    scoring = Scorer,
)

for i, score in enumerate(Scores, start=1):
    print(f"Fold {i}: F1-macro = {score:.4f}")

print(f"\nMean F1-macro: {Scores.mean():.4f}")
print(f"Std  F1-macro: {Scores.std():.4f}")
```

### Join Data Modeling

```python
df = pd.concat(
    [task2[Dub_cols_23], task3[Dub_cols_23]],
    axis = 0,
    ignore_index = True
)
```

```python
df.drop(columns = ["id", "player_id", "platform"], inplace = True)
df = df.dropna(subset = ["segment"])
df["segment"] = df["segment"].astype("int64")
```

```python
X = df.drop("segment", axis = 1)
y = df["segment"]

Classifier = lgb.LGBMClassifier(
    objective = "multiclass",
    num_class = 4,
    random_state = 42,
)

Stratified_K_Fold = StratifiedKFold(n_splits = 5, shuffle = True, random_state = 42)
Scorer = make_scorer(f1_score, average = "macro")

Scores = cross_val_score(
    Classifier,
    X,
    y,
    cv = Stratified_K_Fold,
    scoring = Scorer,
)

for i, score in enumerate(Scores, start = 1):
    print(f"Fold {i}: F1-macro = {score:.4f}")

print(f"\nMean F1-macro: {Scores.mean():.4f}")
print(f"Std  F1-macro: {Scores.std():.4f}")
```
